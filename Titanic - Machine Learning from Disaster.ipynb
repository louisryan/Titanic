## Missing data

Handling missing data is an essential step in data analysis - For many reasons, data often comes incomplete and messy. Therefore, handling missing data is critical. Methods of imputation include:

* Dropping observations or columns.
* Filling with mean, median, previous value etc..
* Replace with values that come from similar records in our dataset that aren't null.

We're going to use a combination of these for our analysis. First, let's get a visual representation of the missing data:

```python
def missing_data(df, format=0):
    # Amend to handle case when no missing data - Does not fail gracefully!
    df_lng = pd.melt(df)
    null_variables = df_lng.value.isnull()
    if format == 0:
        return pd.crosstab(df_lng.variable, null_variables)[pd.crosstab(df_lng.variable, null_variables)[1]>0]
    else:
        nulls = pd.crosstab(df_lng.variable, null_variables)[pd.crosstab(df_lng.variable, null_variables)[1]>0]
        fig = plt.figure(figsize(8,4.5), dpi=1400)
        nulls.plot(kind='barh', stacked=True, legend=False, title = 'Missing Data', alpha=a_bar) 
        
missing_data(train, format=1)
```

On immediate inspection, we can see that the majority of cabin information is missing. This is unfortunate as you would assume there is correlation between cabin location and survival rate. We can come back to this... For now, we'll drop Cabin from our dataframe. 

### Imputation of Data

The next step involved in the analysis is to impute some of the missing data. For the missing data in our dataset, we're going to replace it with values that come from similar records in our dataset that aren't null. For missing ages, we'll use the median age for all passengers. For missing embarked observations, we'll simply use the most common embarkment point.

```python
train = train.drop(['Cabin'], axis=1)

median_age = train['Age'].median()
train['Age'] = train['Age'].fillna(median_age)

embarked_common = train['Embarked'].value_counts().idxmax()
train['Embarked'] = train['Embarked'].fillna(embarked_common)
```

## Random Forests

The Random Forest is a highly versatile learning method with numerous applications ranging from marketing to healthcare and insurance. It can be used to model the impact of marketing on customer acquisition, retention and churn or to predict disease risk and susceptibility in patients.

Random forest is capable of regression and classification. It can handle a large number of features, and it's helpful for estimating which or your variables are important in the underlying data being modeled.

### What is a Random Forest

Random Forest is a solid choice for nearly any prediction problem (even non-linear ones). It's a relatively new machine learning method that came out of Bell Labs in the 90s. It belongs to a larger class of machine learning algorithms called ensemble methods.

Ensemble learning involves the combination of several methods to solve a single prediction problem. It works by generating multiple classifiers/models which learn and make predictions independently. Those predictions are then combined into a single (mega) prediction that should be as good or better than the prediction made by any one classifer. In the case of ensemble methods, the algorithm is combining a series of decision trees that offer often times meaningless results. 

It turns out that even if 99% of the decision trees are meaningless, the 1% of really good observations. The 99% of irrelevant trees cancel each other out and the remainder that are good top that noise and yield good prediction.

### Data Preparation

Before any machine learning can be applied on the data, we must first convert the data to the correct format. scikit-learn accepts a float matrix of m X n shape.

```python
from sklearn import preprocessing as pre
from sklearn_pandas import DataFrameMapper
from sklearn.ensemble import RandomForestClassifier

# encode the appropriate columns
mapper = DataFrameMapper([
                          (['Sex'], pre.LabelEncoder()),           
                          (['Embarked'], pre.LabelEncoder()),
                          (['Age', 'Fare'], pre.StandardScaler())
                          ])
train_np = np.round(mapper.fit_transform(train), 2)

# Split the data into features(X) and target(y) arrays 
X = np.column_stack((train_np, train[['Pclass', 'SibSp', 'Parch']].values))
y = train['Survived'].values

print("The feature set shape is %d rows and %d columns" %(X.shape))
print("The target variable is %d rows and 1 column" %(y.shape))

feature_cols = np.array(['Sex', 'Embarked', 'Age', 'Fare', 'Pclass', 'SibSp', 'Parch'])
```

### Feature Importance

We're going to use scikit-learn to determine which features hold the most predictive power in our data. In predictive modelling, it is best to start with a wide variety of features, determine the important features, remove irrelevant ones and incrementally improve your model by feature engineering.

To do this, we're going to use Random Forests. As the decision trees are randomly generated, the algorithm takes turns leaving out each variable in fitting the model. This allows the Random Forest to calculate how much worse the model does when each variable is left out. The relative rank (i.e. depth) of a feature used as a decision node in a tree can be used to assess the relative importance of that feature with respect to the predictability of the target variable.

```python
rfclf=RandomForestClassifier( max_features='log2'
                             ,n_estimators=500
                             ,min_samples_split=1
                             ,compute_importances=True
                             ,n_jobs=-1).fit(X,y)

importances=rfclf.feature_importances_
sorted_idx=np.argsort(importances)
```

```python
padding=np.arange(X.shape[1])+0.5

pl.barh(padding,importances[sorted_idx],align='center', alpha=a_bar)
pl.yticks(padding,feature_cols[sorted_idx])
pl.xlabel("Relative Importance")
pl.title("Variable Importance")
```

The most relevant features are both Age and Sex with near equal relative importance. Decisions can be made with regard to selecting a subset of features based on this information. We'll first test how our model fits before iterating on the features.

### Fitting and Evaluation Model

Because the dataset is so small, we'll split the data into a training and test set. We'll fit the model on the training data and validate the model on the test data. This is to ensure that we don't overfit the model on the data.

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=0)

print("train shape: %r, train target shape: %r" %(X_train.shape, y_train.shape))
print("test shape: %r, test target shape: %r" %(X_test.shape, y_test.shape))
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='rbf').fit(X_train, y_train)
train_score = svc.score(X_train, y_train)
print('The accuracy of the model on the training data using train_test_split is: %r ' %train_score)
```

```python
pred = svc.predict(X_test)           #Use for confusion matrix
test_score = svc.score(X_test, y_test)
print('The accuracy of the model on the test data is: %r ' %test_score)
```

This actually isn't too bad. The model generated does not generalize well to new, unseen data but not by a significant factor. 

* When the _test_ data score is not as good as the train score, the model is overfitting.
* When the train score is not close to 100% accuracy, the model is underfitting.

Ideally we want to neither overfit nor underfit: test_score ~= train_score ~= 1.0. In our case, we have not learnt our training data well enough meaning the model needs more tuning. 

### Cross Validation

Cross Validation is a procedure to repeat the train/test split several times to get a more reliable and accurate real test score by averaging the values found of the individual runs. The available data is split into two sets. Modelling of the data uses one part. The model created from that data is then tested on the already known data. This process is repeated multiple times using difference splits of the data.

```python
from sklearn.cross_validation import cross_val_score, ShuffleSplit
from scipy.stats import sem

svc = SVC(kernel='rbf').fit(X, y)
cv = ShuffleSplit(len(y), n_iterations=100, test_size=0.2, random_state=0)

# This will produce the test score for each iteration of cross-val
%time test_score = cross_val_score(svc, X,y, cv=cv, n_jobs=-1)

# Utility function to aggregate scores in SEM error
def mean_score(scores):
    return ('Mean Score: {0:.3f} (+/-{1:.3f})').format(np.mean(scores), sem(scores))

mean_score(test_score)
```

### Evaluation

Up to this point, we have used accuracy as the means to test the models performance. Classification accuracy is a popular metric but unfortunately, it's often too simplistic for real world applications. The first additional evaluater we'll use to see how the model is performing is the confusion matrix.

A confusion matrix seperates our rhe decisions made by the classifier, making explicit how one class is being confused for another.

```python
labels = ['Survived', 'Died']

def confusion_matrix(y_test, pred, labels):
    
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(y_test, pred)
    print(cm)
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(cm)
    pl.title('Confusion matrix of the classifier')
    fig.colorbar(cax)
    ax.set_xticklabels([''] + labels)
    ax.set_yticklabels([''] + labels)
    pl.xlabel('Predicted')
    pl.ylabel('True')
    pl.show()

confusion_matrix(y_test, pred, labels)
```

```python
from sklearn import metrics
print metrics.classification_report(y_test, pred, target_names=['Survived', 'Died'])
```

Our model is predicting survivors well but seems to be predicting survival when in reality, they died. In real world applications, this may not be acceptable. If you are building a churn model which attempts to identify if a customer will churn, it may be more acceptable to have a model that predicts a customer will churn when in reality he won't. A worse situation is  where the model predicts a customer will not churn and they end up churning. 

### HyperParameter Optimisation with Grid Search

Cross Validation makes it possible to evaluate the performance of a model class and its hyperparameters on the task at hand. Grid Search attempts to find the optimal weighting of hyperparameters for the selected model. We can print the optimal weights found for each hyperparameter below:

```python
from sklearn.grid_search import GridSearchCV
from pprint import pprint

svc_params = {
              'C' : np.logspace(-1, 2, 4),
              'gamma' : np.logspace(-4, 0, 5)
              }

grid = GridSearchCV(SVC(), svc_params, cv=100, scoring='precision')
%time _ = grid.fit(X, y)
print str.format("The optimal C paramtere is: {0[C]}. The optimal gamma parameter is: {0[gamma]}", grid.best_params_)
```
